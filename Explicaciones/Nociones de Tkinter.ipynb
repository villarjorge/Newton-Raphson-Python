### Programa de hello world

En tkinter primero se declara el elemento (una label por ejemplo) y luego se inicializa. Una label es un bloque al que se le puende poner un atributo de texto.

```python
ventana = Tk() # Declaramos una ventana principal, En ella estarán todos los demás elementos

mi_etiqueta = Label(master=ventana, text="Hello world") # Creamos la etiqueta (Declaramos). Le estamos diciendo al programa que 
# queremos mi_etiqueta en ventana y que queremos que tenga de texto "Hello world"

mi_etiqueta.pack() # Metemos la etiqueta en la ventana (Inicializamos)

ventana.mainloop() # Esto hace que el código que hemos escrito se repita múltiples veces hasta que se cierre la ventana
```

### Sistema de posicionamiento .grid

En el programa anterior utilizamos .pack, que es el sistema más simple que hay. .pack() coloca los elementos donde hay espacio disponible. La mayoría del tiempo utilizaremos .grid() dado que permite mayor control. .grid() permite colocar los elementos en fución de una cuadrícula con columnas y filas a especificar (empezando en cero). Un ejemplo:

```python
ventana = Tk() 
# Declaramos
mi_etiqueta1 = Label(master=ventana, text="Hello world") 
mi_etiqueta2 = Label(master=ventana, text="Hola mundo")
# Inicializamos
mi_etiqueta1.grid(row=0, column=0) # row->fila, column->columna
mi_etiqueta2.grid(row=1, column=1) # Se colocarán relativamente de el uno al otro. Tkinter ignora las columnas vacías, por lo
# que si pones un número muy grande el progama no cambiará

ventana.mainloop() # Esto hace que el código que hemos escrito se repita múltiples veces hasta que se cierre la ventana
```

### Botones

Otro tipo de elemento como la etiqueta

```python
ventana = Tk() 

mi_boton = Button(master=ventana, text="Soy un boton") # De la misma forma le estamos diciendo al programa que queremos mi_boton
                                                       # en ventana y que queremos que tenga el texto "Soy un boton"
# los botones tienen otros atributos, muchos de ellos compartidos con label y con otros (Recuerda utilizar help()):
mi_boton1 = Button(master=ventana, text="No me puedes tocar", state=DISABLED)
mi_boton2 = Button(master=ventana, text="Soy más grande", padx=50, pady=50) # padx y pady aumentan el tamaño en la x y en la y 
# respectivamente

mi_boton.pack()
mi_boton1.pack()
mi_boton2.pack()

ventana.mainloop()
```

#### Como hacer que los botones hagan cosas

Tenemos que hacer una función, en este caso será una muy simple.

```python
ventana = Tk() 

def Click(): # cuando el boton sea presionado se llamará a esta función y se mostrará en pantalla un mensaje 
    mi_etiqueta = Label(master=ventana, text="Has hecho click!")
    mi_etiqueta.pack()

mi_boton = Button(master=ventana, text="Tengo funcionalidad", command=Click) # El último atributo command es lo que hace que
# cuando se presione el botón se llame a la función y por tanto se ejecute el código de ella. Si pones paréntesis despues de la 
# función no funcionará bien

mi_boton.pack()

ventana.mainloop()
```

### Accediendo al texto introducido por el usuario

Utilizamos .get() para tomar el texto de una entrada. En este ejemplo utilizaremos otro ejemplo anterior para crear una etiqueta con una string introducida por el usuario. También podemos utilizar .insert() para poner un texto por defecto

```python
ventana = Tk() 

entrada = Entry(master=ventana)
entrada.pack()
entrada.insert(0, "Introduze tu nombre: ")

def Click(): 
    mi_etiqueta = Label(master=ventana, text=f"Hola {entrada.get()}") # Ahora la función hará que aparezca en pantalla 
    # el nómbre introducido en la entrada de texto
    mi_etiqueta.pack()

mi_boton = Button(master=ventana, text="Introduce tu nombre", command=Click)

mi_boton.pack()

ventana.mainloop()
```

## Una calculadora básica

Utilizaremos lo aprendido para crear una calculadora que tenga diez botones para los números, uno para borrar, uno para realizar la suma... 

```python
ventana = Tk()
ventana.title("Calculadora simple") # Esto controla el título de nuesto programa

entrada = Entry(master=ventana, width=35, borderwidth=5) # Esta entrada será donde se verán los números clicados
entrada.grid(row=0, column=0, padx=5, pady=10, columnspan=3) # El último parámetro hace que el elemento tome tres columnas,
#  lo queremos así para que sea tan ancho como los tres botones que tendrá abajo

# Declaramos una función que pondrá el número clicado en la entrada de texto "entrada"
def numero_presionado(num):
    n_actual = entrada.get()
    entrada.delete(0, END)
    entrada.insert(0, str(n_actual) + str(num))
    
# Declaramos una función que borra lo que halla en "entrada"
def borrar_entrada():
    entrada.delete(0, END)
    
# Declaramos una varaias funciones que guarden los números en función de si se quiere sumar, restar
def recordar_num_suma():
    primer_num = entrada.get()
    global f_num # Creamos una variable global. Esta variable será el primer número introducido por el usuario antes de
                 # presionar "+"
    global simbolo # Esta variable global servira para distinguir que se debe hacer con el número 
    simbolo = "sumar"
    f_num = float(primer_num)
    entrada.delete(0, END)
def recordar_num_resta():
    primer_num = entrada.get()
    global f_num
    global simbolo
    simbolo = "restar"
    f_num = float(primer_num)
    entrada.delete(0, END)
def recordar_num_multiplicación():
    primer_num = entrada.get()
    global f_num 
    global simbolo
    simbolo = "multiplicar"
    f_num = float(primer_num)
    entrada.delete(0, END)
def recordar_num_division():
    primer_num = entrada.get()
    global f_num 
    global simbolo
    simbolo = "dividir"
    f_num = float(primer_num)
    entrada.delete(0, END)
    
# Creamos una función que tomará el número anterior y lo operará al que está en "entrada" en ese momento
def operar_nums():
    segundo_num = float(entrada.get())
    entrada.delete(0, END)
    if simbolo == "sumar":
        entrada.insert(0, f_num + segundo_num)
    elif simbolo == "restar":
        entrada.insert(0, f_num - segundo_num)
    elif simbolo == "multiplicar":
        entrada.insert(0, f_num*segundo_num)
    else:
        entrada.insert(0, f_num/segundo_num)
        
# Creamos los botones de los números. Utilizamos lambda para pasar un parámetro a la función
boton_1 = Button(master=ventana, text="1", padx=40, pady=20, command=lambda: numero_presionado(1))
boton_2 = Button(master=ventana, text="2", padx=40, pady=20, command=lambda: numero_presionado(2))
boton_3 = Button(master=ventana, text="3", padx=40, pady=20, command=lambda: numero_presionado(3))
boton_4 = Button(master=ventana, text="4", padx=40, pady=20, command=lambda: numero_presionado(4))
boton_5 = Button(master=ventana, text="5", padx=40, pady=20, command=lambda: numero_presionado(5))
boton_6 = Button(master=ventana, text="6", padx=40, pady=20, command=lambda: numero_presionado(6))
boton_7 = Button(master=ventana, text="7", padx=40, pady=20, command=lambda: numero_presionado(7))
boton_8 = Button(master=ventana, text="8", padx=40, pady=20, command=lambda: numero_presionado(8))
boton_9 = Button(master=ventana, text="9", padx=40, pady=20, command=lambda: numero_presionado(9))
boton_0 = Button(master=ventana, text="0", padx=40, pady=20, command=lambda: numero_presionado(0))

boton_sumar = Button(master=ventana, text="+", padx=39, pady=20, command=recordar_num_suma)
boton_restar = Button(master=ventana, text="-", padx=41, pady=20, command=recordar_num_resta)
boton_multiplicar = Button(master=ventana, text="*", padx=40, pady=20, command=recordar_num_multiplicación)
boton_dividir = Button(master=ventana, text="/", padx=40, pady=20, command=recordar_num_division)

boton_igual = Button(master=ventana, text="=", padx=91, pady=20, command=operar_nums)
boton_borrar = Button(master=ventana, text="Borrar", padx=72, pady=20, command=borrar_entrada)

# Ponemos botones en la pantalla. Recuerda que "entrada" ya está en la pantalla
boton_1.grid(row=3, column=0)
boton_2.grid(row=3, column=1)
boton_3.grid(row=3, column=2)

boton_4.grid(row=2, column=0)
boton_5.grid(row=2, column=1)
boton_6.grid(row=2, column=2)

boton_7.grid(row=1, column=0)
boton_8.grid(row=1, column=1)
boton_9.grid(row=1, column=2)

boton_0.grid(row=4, column=0)

boton_sumar.grid(row=5, column=0)
boton_restar.grid(row=6, column=0)
boton_multiplicar.grid(row=6, column=1)
boton_dividir.grid(row=6, column=2)

boton_igual.grid(row=5, column=1, columnspan=2)
boton_borrar.grid(row=4, column=1, columnspan=2)

ventana.mainloop()
```

### Utilizando marcos

### Botones radiales

Botones que pueden estar encendidos o apagados. Tienen su propia variable de control, son diferentes de las varialbles de python.

```python
ventana = Tk() 
ventana.title("Botones radiales")

r = IntVar() # Esta es una variable de ktinker que se debe acceder como si fuera un campo de introducción con .get
r.set(1)

def seleccionada_cual(valor): # Esta función es la que se encarga de actualizar todo cuando se marca una opción
    etiqueta = Label(ventana, text=f"Has seleccionado la opción: {valor}") # No hace falta poner el "master=ventana" siempre
    etiqueta.pack()
    
boton_r1 = Radiobutton(master=ventana, text="Opción 1", variable=r, value=1, command=lambda: seleccionada_cual(r.get())) 
# cuando este botón se pulse a la variable "r" se le asignará el valor 1 y se ejecutará la función 
boton_r2 = Radiobutton(master=ventana, text="Opción 2", variable=r, value=2, command=lambda: seleccionada_cual(r.get()))

etiqueta = Label(master=ventana, text=f"Has seleccionado la opción: {r.get()}") # Esta etiqueta mostrará que opción 
# se ha selecionado

boton_r1.pack()
boton_r2.pack()

etiqueta.pack()

ventana.mainloop()
```

### Como utilizar un bucle para generar elementos

Utilizaremos una lista y un bucle para reducir la longitud de nuestro código. En este caso será para hacer un menú de un restaurante de pizzas.

```python
ventana = Tk() 
ventana.title("Pizzas")

TIPOS = [ # Esta lista tendrá como queremos que se llame el botón y como queremos que salga al hacer click en enviar
    #("Texto para el boton", "Texto para enviar")
    ("Pepperoni", "pepperoni"),
    ("Queso", "queso"),
    ("Champiñon", "champiñones"),
    ("Cebolla", "cebolla"),
]

pizza = StringVar()
pizza.set("Pepperoni")

for texto_boton, tipo in TIPOS:
    Radiobutton(ventana, text=texto_boton, variable=pizza, value=tipo).pack(anchor=W)  
    # En vez de declararlo como una variable y luego ponerlo en pantalla simplemente lo ponemos en pantalla. 
    # El argumento de pack indica que se debe anclar hacia el "west" oeste, izquierda

def hecho_click(valor):
    etiqueta = Label(ventana, text=f"Pizza de {valor} enviada")
    etiqueta.pack()
    
boton = Button(ventana, text="Enviar", command=lambda: hecho_click(pizza.get()))
boton.pack()

ventana.mainloop()
```

### Crear nuevas ventanas

Creamos nuevas ventanas. Si cierras la ventana secundaria no se cierra la principal pero si cierras la principal se cierran las dos.

```python
#Ejemplo 1
ventana = Tk() 
ventana.title("Ventana 1")

ventana_secundaria = Toplevel()
ventana_secundaria.title("Ventana 2")

etiqueta = Label(ventana_secundaria, text="Estoy en una ventana secundária")
etiqueta.pack(padx=70)

ventana.mainloop()
```

```python
#Ejemplo 2
ventana = Tk() 
ventana.title("Ventana 1")

def abrir_ventana2():
    global etiqueta # Esto es necesário para que se mueste el elemento
    ventana_secundaria = Toplevel()
    ventana_secundaria.title("Ventana 2")
    etiqueta = Label(ventana_secundaria, text="Estoy en una ventana secundária")
    etiqueta.pack(padx=70)
    boton2 = Button(ventana_secundaria, text="Cerrar ventana", command=ventana_secundaria.destroy)
    boton2.pack()
boton = Button(ventana, text="Abrir ventana secundária", command=abrir_ventana2)
boton.pack()

ventana.mainloop()
```
